# Amines for Carbon Capture on AWS-SV1

We evaluate VQE on the AWS-SV1 remote simulator for system $CO_2CH_3NH_2$. We use pennylane for this purpose and it has pennylane-braket-plugin. AWS-S1 device can simulate upto 34 qubits but here we only use 6.

```python
import sys
sys.path.append('..')
```

```python
import pickle 
import pennylane as qml
import pennylane.numpy as np
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
from utils import rel_err
```

## Hamiltonian construction

We can use pennylane for constructing the hamiltonian of the system. However, since we earlier used Qiskit for this purpose, we save the hamiltonian in a pickle file and now load it here. This is important for consistency in the results obtained. PennyLane/Braket pipeline, provides built-in support for pre-grouping the observables in a Hamiltonian to reduce the number of device executions, saving both runtime and simulation costs when employing distant devices.

```python
with open('hamiltonian.pickle', 'rb') as handle:
    h_dict = pickle.load(handle)
```

```python
ops = []

for op in h_dict['ops']:
    ops.append(qml.pauli.string_to_pauli_word(op))

coeffs = []
for coeff in h_dict['coeffs']:
    coeffs.append(coeff)

H = qml.Hamiltonian(coeffs, ops)
H.compute_grouping()
```

```python
len(H.ops)
```

## Simulation on lightning.qubit

```python
qubits = 6
dev = qml.device("lightning.qubit", wires=qubits)
```

```python
@qml.qnode(dev)
def cost_fn(param):
    qml.BasisState(np.array([1,0,1,1,0,1]), wires=range(qubits))
    
    for q in range(qubits):
        qml.RY(param[q][0], q)
    
    for q in range(qubits - 1):
        qml.CNOT((q, q+1))
    
    for q in range(qubits):
        qml.RY(param[q][1], q)
        
    return qml.expval(H)
```

```python
# Define a simple callback function
exact_energy = -155.628847646236
active_space_part = -277.061327364244
nuc_rep = 154.047087551131

intermediate_info = []
five_percent = []
one_percent = []

def callback(value):
    intermediate_info.append(value)
    five_percent.append(exact_energy*(1-0.05))
    one_percent.append(exact_energy*(1-0.01))
    clear_output(wait=True)
    plt.plot(intermediate_info, color='purple', lw=2, label=f'Simulated VQE {np.round(value,4)}')
    plt.ylabel('Energy')
    plt.xlabel('Iterations')
    # Exact ground state energy value
    plt.axhline(y=exact_energy,color="tab:red",ls="--", lw=2,label="Target: "+str(np.round(exact_energy,4)))
    plt.plot(five_percent, lw=1, label=f'5% ({np.round(exact_energy*(1-0.05),4)})')
    plt.plot(one_percent, lw=1, label=f'1% ({np.round(exact_energy*(1-0.01),4)})')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
max_iterations = 100
opt = qml.SPSAOptimizer(maxiter=max_iterations,a = 0.1, c = 0.1, A = 1)
theta = np.array(np.random.uniform(size=(6,2)), requires_grad=True)

# store the values of the cost function
energy = [cost_fn(theta.real)]

conv_tol = 1e-06

for n in range(max_iterations):
    theta, prev_energy = opt.step_and_cost(cost_fn, theta.real)
    callback(prev_energy)
    energy.append(cost_fn(theta.real))
    conv = np.abs(energy[-1] - prev_energy)
    if conv <= conv_tol:
        break

print("\n" f"Final value of the ground-state energy = {energy[-1]:.8f} Ha")
```

```python
# Compute the relative error between the expected ground state energy and the VQE's output
rel_error = rel_err(exact_energy, energy[-1])

print(f'Expected ground state energy: {exact_energy:.12f}')
print(f'Computed ground state energy: {energy[-1]:.12f}')
print(f'Relative error: {rel_error:.12f}')
```

## Deparameterisation

```python
def cost_fn_deparam(param):
    qml.BasisState(np.array([1,0,1,1,0,1]), wires=range(qubits))
    
    qml.RY(param[0], 0)
    qml.RY(np.pi/2, 1)
    qml.RY(-np.pi/2, 2)
    qml.RY(-np.pi/2, 3)
    qml.RY(3*np.pi/2, 4)
    qml.RY(np.pi/2, 5)
    
    for q in range(qubits - 1):
        qml.CNOT((q, q+1))
    
    qml.RY(param[1], 0)
    qml.RY(-np.pi/2, 1)
    qml.RY(np.pi/2, 2)
    qml.RY(3*np.pi/2, 3)
    qml.RY(3*np.pi/2, 4)
    qml.RY(3*np.pi/2, 5)
        
    return qml.expval(H)
```

```python
dev = qml.device('lightning.qubit', wires = qubits)
cost_fn_deparam_local = qml.QNode(cost_fn_deparam, device = dev)
cost_fn_deparam_local
```

```python
max_iterations = 10
opt = qml.SPSAOptimizer(maxiter=max_iterations,a = 0.01, c = 0.01, A = 1)

initial_point = [1.9192679, 4.61095253]
theta_deparam = np.array(initial_point, requires_grad=True)

# store the values of the cost function
for n in range(max_iterations):
    theta_deparam, prev_energy = opt.step_and_cost(cost_fn_deparam_local, theta_deparam.real)
    callback(prev_energy)

computed_energy = prev_energy.real
print("\n" f"Final value of the ground-state energy = {computed_energy:.12f} Ha")
```

```python
# Compute the relative error between the expected ground state energy and the VQE's output
rel_error = rel_err(exact_energy, computed_energy)

print(f'Expected ground state energy: {exact_energy:.12f}')
print(f'Computed ground state energy: {computed_energy:.12f}')
print(f'Relative error: {rel_error:.12f}')
```

## Simulation on default.mixed device with ZNE error mitigation

```python
def _cost_fn_deparam(param):
    
    qml.PauliX(0)
    qml.PauliX(2)
    qml.PauliX(3)
    qml.PauliX(5)
    
    qml.RY(param[0], 0)
    qml.RY(np.pi/2, 1)
    qml.RY(-np.pi/2, 2)
    qml.RY(-np.pi/2, 3)
    qml.RY(3*np.pi/2, 4)
    qml.RY(np.pi/2, 5)
    
    for q in range(qubits - 1):
        qml.CNOT((q, q+1))
    
    qml.RY(param[1], 0)
    qml.RY(-np.pi/2, 1)
    qml.RY(np.pi/2, 2)
    qml.RY(3*np.pi/2, 3)
    qml.RY(3*np.pi/2, 4)
    qml.RY(3*np.pi/2, 5)
        
    return qml.expval(H)
```

```python
# Describe noise
noise_gate = qml.DepolarizingChannel
noise_strength = 0.05

# Load devices
dev_ideal = qml.device("default.mixed", wires=qubits)
dev_noisy = qml.transforms.insert(noise_gate, noise_strength, position="all")(dev_ideal)

qnode_noisy = qml.QNode(_cost_fn_deparam, dev_noisy)
qnode_ideal = qml.QNode(_cost_fn_deparam, dev_ideal)
```

```python
print(qml.draw(qnode_ideal)((theta_deparam)))
```

```python
from pennylane.transforms import mitigate_with_zne

scale_factors = [1, 2, 3]

qnode_mitigated = mitigate_with_zne(
    scale_factors=scale_factors,
    folding=qml.transforms.fold_global,
    extrapolate=qml.transforms.richardson_extrapolate,
)(qnode_noisy)

print("Ideal QNode: ", qnode_ideal(theta_deparam.real))
print("Mitigated QNode: ", qnode_mitigated(theta_deparam.real))
print("Noisy QNode: ", qnode_noisy(theta_deparam.real))
```

## The Remote Braket Simulator

We use the optimal parameters obtained from the deparameterisation as the initial point and execute the circuit on the SV1 simulator.

```python
# Use Braket SDK Cost Tracking to estimate the cost to run this example
from braket.tracking import Tracker
tracker = Tracker().start()
```

```python
device_arn = 'arn:aws:braket:::device/quantum-simulator/amazon/sv1'
```

```python
dev = qml.device('braket.aws.qubit', device_arn=device_arn, wires=qubits)
cost_fn_braket_remote = qml.QNode(_cost_fn_deparam, device=dev)
cost_fn_braket_remote
```

```python
scale_factors = [1, 2, 3]

qnode_mitigated = mitigate_with_zne(
    scale_factors=scale_factors,
    folding=qml.transforms.fold_global,
    extrapolate=qml.transforms.richardson_extrapolate,
)(cost_fn_braket_remote)
```

```python
computed_energy = qnode_mitigated(theta_deparam.real)
print("\n" f"Final value of the ground-state energy = {computed_energy:.12f} Ha")
```

```python
tracker.stop()
```

```python
tracker.simulator_tasks_cost()
```

```python
tracker.quantum_tasks_statistics()
```

```python
# Compute the relative error between the expected ground state energy and the VQE's output
rel_error = rel_err(exact_energy, computed_energy)

print(f'Expected ground state energy: {exact_energy:.12f}')
print(f'Computed ground state energy: {computed_energy:.12f}')
print(f'Relative error: {rel_error:.12f}')
```

```python
total_energy = computed_energy + active_space_part + nuc_rep

ev = 27.2114
actual_energy = -279.029888426681*ev


print(f'Actual total energy: {actual_energy:.12f} Ha\n')
print(f"Total energy: {total_energy:.12f} Ha")
print(f"Total energy in ev: {total_energy*ev:.12f} ev\n")

rel_error = rel_err(actual_energy, total_energy*ev)
print(f'Relative error: {rel_error:.12f}')
```

## Results

The circuit execution takes ~1500 seconds with a relative error within $10^{-3}$ Hartree.

The relative error of `VQE's ground state energy` with the `classical reference solution` is within $10^{-2}$ Hartree.
